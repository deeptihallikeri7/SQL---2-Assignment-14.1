```python
import sqlite3
import time
from numpy import genfromtxt

def dict_factory(cursor, row):
    d = {}
    for idx, col in enumerate(cursor.description):
        d[col[0]] = row[idx]
    return d
```

```python
def Create_DB(db):      
    #Create DB and format it as needed
    with sqlite3.connect(db) as conn:
        conn.row_factory = dict_factory
        conn.text_factory = str

        cursor = conn.cursor()
        create_Table_Query = "CREATE TABLE [ADULT_Details] ([id] INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL UNIQUE,"
        create_Table_Query = create_Table_Query + " [AGE] INTEGER, [WORKCLASS] Text, [final_weight] FLOAT, [EDUCATION] Text,[EDUCATION_NUM] Text,"
        create_Table_Query = create_Table_Query + " [MARTIAL_STATUS] Text,[OCCUPATION] Text, [RELATIONSHIP] Text, [RACE] Text,[SEX] Text, [CAPITAL_GAIN] INTEGER,"
        create_Table_Query = create_Table_Query + " [CAPITAL_LOSS] Text,[HOURS_per_Week] Text, [NATIVE_COUNTRY] Text, [SALabove50K] Text)"
        cursor.execute(create_Table_Query)
  
```

```python
def Add_Record(db, data):
    #Insert record into table
    with sqlite3.connect(db) as conn:
        print("Starting Add Record...")
        conn.row_factory = dict_factory
        conn.text_factory = str
        cursor = conn.cursor()
        cursor.execute("""INSERT INTO ADULT_Details({cols}) VALUES({vals});""".format(cols = (str(data.keys()).strip('[]')), vals = (str([data[i] for i in data]).strip('[]'))))
        print("Ending Add Record...")
```

```python
def Loading_Data(file_name):
    data = genfromtxt(file_name, delimiter=',') #, skiprows=1,converters={0: lambda s: str(s)})
    return data.tolist()
```

```python
db = 'csv_test_sql.db' #Database filename 
file_name = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data" 
print("Loading_Data.... ")
data = Loading_Data(file_name) #Get data from CSV
```

```python
def Drop_DB(db):      
        with sqlite3.connect(db) as conn:
            conn.row_factory = dict_factory
            conn.text_factory = str
            cursor = conn.cursor()
            Drop_Table_Query = "drop table if exists [ADULT_Details];"
            cursor.execute(Drop_Table_Query)
```

```python
Drop_DB(db)
Create_DB(db) #Create DB

#For every record, format and insert to table
for i in data:
    record = {
#col_Names = ['AGE', 'WORKCLASS', 'final_weight', 'EDUCATION','EDUCATION_NUM', 'MARTIAL_STATUS','OCCUPATION', 'RELATIONSHIP',
#'RACE','SEX', 'CAPITAL_GAIN', 'CAPITAL_LOSS','HOURS_per_Week', 'NATIVE_COUNTRY', 'SALabove50K']
                'AGE' : i[0],
                'WORKCLASS' : i[1],
                'final_weight' : i[2],
                'EDUCATION' : i[3],
                'EDUCATION_NUM' : i[4],
                'MARTIAL_STATUS' : i[5],
                'OCCUPATION' : i[6],
                'RELATIONSHIP' : i[7],
                'RACE' : i[8],
                'SEX' : i[9],
                'CAPITAL_GAIN' : i[10],
                'CAPITAL_LOSS' : i[11],
                'HOURS_per_Week' : i[12],
                'NATIVE_COUNTRY' : i[13],
                'SALabove50K' : i[14]
            }
    Add_Record(db, record)
```
